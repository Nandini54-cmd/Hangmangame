```python
import random
import hangman_stages
import word_file

def hangman_game():
    lives = 6  # Number of lives
    chosen_word = random.choice(word_file.words)  # Randomly choose a word from the list
    display = ['-'] * len(chosen_word)  # Initialize the display as a list of hyphens ('-')
    guessed_letters = []  # List to track guessed letters
    game_over = False

    print("Welcome to Hangman!")

    while not game_over:
        # Display the current state of the word and number of lives
        print(f"Current word: {' '.join(display)}")
        print(f"Lives remaining: {lives}")
        print(hangman_stages.stages[6 - lives])  # Show current hangman stage

        # Prompt user for a guess
        guessed_letter = input("Guess a letter: ")

        # Check if the user input is valid
        if len(guessed_letter) != 1 or not guessed_letter.isalpha():
            print("Please enter a single letter.")
            continue

        if guessed_letter in guessed_letters:
            print(f"You've already guessed '{guessed_letter}'. Try again.")
            continue

        guessed_letters.append(guessed_letter)

        # Check if the guessed letter is in the chosen word
        if guessed_letter in chosen_word:
            for position in range(len(chosen_word)):
                if chosen_word[position] == guessed_letter:
                    display[position] = guessed_letter  # Update the display list
            print(f"Good guess! '{guessed_letter}' is in the word.")
        else:
            lives -= 1  # Decrease lives for incorrect guess
            print(f"Sorry, '{guessed_letter}' is not in the word.")

        # Check if the player has won (i.e., there are no more hyphens in display)
        if '-' not in display:
            print(f"Congratulations! You've guessed the word: {chosen_word}")
            game_over = True

        # Check if the player has lost (i.e., lives reached 0)
        if lives == 0:
            print(f"Game Over! The word was: {chosen_word}")
            game_over = True

if __name__ == "__main__":
    hangman_game()
```
